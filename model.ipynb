```python
import pandas as pd

# Read in the data
df_aeb = pd.read_csv('..\\SleepHealthDataTables\\active-energy-burned.csv')
df_aet = pd.read_csv('..\\SleepHealthDataTables\\apple-exercise-time.csv')
df_beb = pd.read_csv('..\\SleepHealthDataTables\\basal-energy-burned.csv')
df_bm = pd.read_csv('..\\SleepHealthDataTables\\body-mass.csv')
df_dwr = pd.read_csv('..\\SleepHealthDataTables\\distance-walking-running.csv')
df_fcl = pd.read_csv('..\\SleepHealthDataTables\\flights-climbed.csv')
df_hr = pd.read_csv('..\\SleepHealthDataTables\\heart-rate.csv')
df_h = pd.read_csv('..\\SleepHealthDataTables\\height.csv')
df_nt = pd.read_csv('..\\SleepHealthDataTables\\nap-tracker.csv')
df_sqc = pd.read_csv('..\\SleepHealthDataTables\\sleep-quality-checker.csv')
df_sc = pd.read_csv('..\\SleepHealthDataTables\\step-count.csv')


# Make a list of the dataframes
dfs = [df_aeb, df_aet, df_beb, df_bm, df_dwr, df_fcl, df_hr, df_h, df_nt, df_sqc, df_sc]

# Make a list of Strings with the names of the dataframes
df_names = ['active_energy_burned', 'apple_exercise_time', 'basal_energy_burned', 'body_mass', 'distance_walking_running', 'flights_climbed', 'heart_rate', 'height', 'nap_tracker', 'sleep_quality_checker', 'step_count']

# Sort each dataframe by participantId
for df in dfs:
    df.sort_values('participantId', inplace=True)

# merged_df = reduce(lambda left,right: pd.merge(left,right,on='participantId', how='outer'), dfs)

# print(merged_df.head())


# Create summary statistics for each dataframe for value column
for idx, curr_df in enumerate(dfs, start=1):
    # Print the name of the dataframe
    print(f"DataFrame {idx} Name: {df_names[idx-1]}")
    if 'value' in curr_df.columns:
        print(curr_df['value'].describe())
    elif 'sq_score' in curr_df.columns:
        print(curr_df['sq_score'].describe())
    elif 'NapQuality' in curr_df.columns:
        print(curr_df['NapQuality'].describe())
        print(curr_df['NapDuration'].describe())
```
